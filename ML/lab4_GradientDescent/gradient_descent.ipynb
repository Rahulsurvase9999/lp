```
def cost_function(x):
    return (x+3)**2
```

```
def costFunc_deriv(x):
    return 2*(x+3)
    
```

```
def gradient_descent(initial_x, learning_rate,no_iteration):
    x=initial_x
    for i in range(no_iteration):
        gradient = costFunc_deriv(x)
        x = x - learning_rate * gradient
        print(f"Iteration {i+1} : x={x} , y = {cost_function(x)}")
    return x
```

```
initial_x = 2
learning_rate = 0.1
no_iteration = 120
```

```
local_minima = gradient_descent(initial_x,learning_rate,no_iteration)
```

```
print(f"Local minimum occurs at x = {local_minima}, y = {cost_function(local_minima)}")
```
